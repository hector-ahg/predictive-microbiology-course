## Conditional expressions, loops and functions

To begin to build up more complex programs, we will need to get to grips with three important programming concepts:

 * At an intermediate point through a program, we may wish to execute different operations depending on its internal state without providing external intervention. To distinguish different states of the program and define the resulting sequence of operations in each case, we use **conditional expressions**.
 * In many cases, we wish to apply a particular sequence of operations multiple times. Explicitly writing out each of these repetitions would be cumbersome, and in many cases we don't know how many times to repeat the operations until we run the code. To get around these issues, we use **loops**.
 * Similarly, we may wish to apply the same sequence of operations to multiple different sets of data. Rather than explicitly write out the full set of operations each time we wish to apply them, we usually package them up into **functions**.

### Conditional expressions

Conditional expressions are the main means by which we control the flow of a program. Python has only a single type of conditional expression, the **if** statement, which is implemented almost as if writing a phrase in English:

````python
a = -3

if a < 0:
    print('a is negative') # VERY IMPORTANT! Note indentation (white space before the code statements).
else:
    print('a is positive')
````

The expression a < 0 evaluates to a Boolean, in this case True. This result is then used by the if statement to determine which of the following code blocks should be executed: if True, the block immediately following the 'if ...:' line is executed. If False, the block following the 'else:' expression is evaluated instead.

If we wish to distinguish between more than two possibilities, we can make use of the 'elif ...:' statement:

````python
b = 0

if b < 0:
    print('b is negative')
elif b == 0: # Note the use of the double equals sign (==), which returns True if both sides are equal.
    print('b equals 0')      #This differs from the use of the single =, which is used as the variable assignment operation.
else:
    print('b is positive')
````

In this case, the initial condition check (b < 0) returns False, so control flows to the next condition check (b == 0), which returns True and causes the following code block to be executed. All subsequent condition checks (in this case, the 'else:' line) are skipped.

### For loops

For loops are used when we know how many times we wish to repeat an block of code before we execute that section of the program. As an example, one (inefficient) way to multiply 2 by 4 is to add 2 to itself 4 times:

````python
for x in [1,2,3]:
    print(x)
````

````python
a = 2
b = 4

total = 0

for i in range(b): #i is incremented during each cycle of the loop, going through the sequence i = 0, i = 1, i = 2, ... i = (b-1)
    print(i)
    total += a #The notation x += y is equivalent to x = x + y.

print(total)
````

### While loops

While loops are used when we *do not* know how many times we wich to repeat a code block before we come to execute that section of the code. For example, one way of performing division is to count how many times you must subtract one number from another before it equals 0:

````python
a = 3
b = 18

count = 0

while b > 0: #The operation b > 0 returns a Boolean which is True as long as b is greater than 0. As soon as it switches to False, the loop terminates.
    b -= a
    count += 1

print(count)
````

While loops can be dangerous, as they can result in the system getting stuck if the condition they are checking can never be false. The most basic example of this would be to simply write 'while True:'.

### Function definitions

If we want to use a particular piece of code in multiple places, it is often convenient to package that piece of code up as a function. Functions take in a given number of 'arguments' and manipulate them internally, resulting in one or many outputs. We can very easily turn our multiplication example into a function:

````python
def multiply(int1,int2): #int1 and int2 are inputs
    total = 0

    for i in range(int2):
        total += int1

    return total #return operation defines the output of the function
````

````python
print(multiply(3,4))
print(multiply(9,2))

#Functions can also take in variables
a = 5 # define a variable a
b = 20 # deine a variable b
print(multiply(a,b))
````

An important idea to note here is the notion of **scope**. The scope of a particular name binding (i.e. what the variable is called) represents the part of the program in which that binding is valid.

In this case, we can see that the scope of the variables int1 and int2 is restricted to the multiply function definition - if for example we were to call other variables int1 and int2 in a different part of the code, this would have no effect on the output of this function:

````python
print(multiply(3,5)) #Outputs 15

int1 = 24 # here, int1 is a variable, independent from the function multiply(int1,int2) defined above.
int2 = 15 # same for int2

print(multiply(3,5)) #Still outputs 15
````

### Class definition

In Python, a class is a blueprint or template for creating objects. It allows you to group data (attributes) and behaviours (methods) into a single logical unit. For example, a "Car" class might specificy that every car has a color and can drive(). The constructor __init__ is an special method that runs automatically when you create a new object. Its main job is to set the initial state (attrributes) of the object.
The self parameter is mandatory first argument for methods within a class. It refers to the specific instance of the object being worked on, allowing Python to distinguish between differenrt objects of the same class.  

Use Class for complex workflows

````python
# Define class Bacteria with attributes (name, shape, gram_stain) and method (describe itself))
class Bacteria:
    microorganism_type = "Prokaryote"
    def __init__(self, name, shape, gram_stain):
        self.name = name
        self.shape = shape
        self.gram_stain = gram_stain
    def describe(self):
        return f"{self.name} is a {self.shape}-shaped, {self.gram_stain} bacterium."
````

### The ndarray object

The core element of NumPy is the ndarray, which stands for n-dimensional array. Arrays are storage for numerical data, and include vectors (1-dimensonal arrays) and matrices (2-dimensional arrays).

Data is assigned to and read out from arrays by specifying the coordinates of the locations you want to access. The **first coordinate refers to the rows and the second to columns**:

````python
testData = np.zeros((4,4)) #Initialise a 4x4 matrix (ndarray) filled with zeros

print(" ") # Separators
print(testData)
print(" ")

# Modify matrix

testData[1,2] = 3 #Assign the value 3 to row 2, column 3

print(" ")
print(testData) #Print the whole array
print(" ")
print(testData[1,2]) #Print just the modified element
print(" ")
````

You can also access multiple elements at once by specifying a range of elements in each dimension:

````python
testData2 = np.zeros((4,4))

testData2[:,1] = 1 #The : indicates that all rows should be accessed

print(testData2)
print(' ')

testData2[2,:2] = 2 #Specify that the first two columns of row 2 should be accessed

print(testData2)
print(' ')

testData2[0,-1] = 3 #Can use negative indicies to count from the end of the array

print(testData2)
````

Notice that Python starts counting from 0 like the Mayans!

### Manipulating arrays

NumPy provides a great many tools for manipulating this basic ndarray object. The following code illustrates some common examples - for more, see the [NumPy documentation](https://numpy.org/doc/stable/reference/index.html#reference). In general, if you can think of some way of reshaping or otherwise manipulating an array, there will be some function within NumPy that can do it already.

````python
seqLst = np.arange(5) #Create a 1D ndarray containing a sequence of numbers

print(seqLst)
print(' ')

seqArray1 = np.tile(seqLst, (5,1)) #Repeat the sequence array 5x as a row

print(seqArray1)
print(' ')

seqArray2 = np.tile(seqLst, (1,5)) #Repeat the sequence array 5x as a column

print(seqArray2)
print(' ')

transArray = seqArray1.T #The transpose of the matrix represented by seqArray1. Trenspose means: rows become columns and viceversa.

print(transArray)
print(' ')

multArray = np.multiply(seqArray1,transArray) #The elementwise product of seqArray1 and its transpose

print(multArray)
print(' ')

dotProduct = np.dot(seqArray1,transArray) #The dot product multipies rows by columns and add them
print(dotProduct)
````

## Pythonic snake pits

As with any language, Python can sometimes behave in a counterintuitive way. Usually, these behaviours represent something fundamental about the way the language works, and can lead to a deeper understanding of the subject. However, at other times they can just be irritating.

There are some good articles describing some of these quirks (intermediate users may wish to see [this article](https://deepsource.io/blog/python-common-mistakes/) for an example). One particularly problematic trip hazard for users coming from other languages is the way that variable assignment works:

````python
lstA = [1,2,3,4]
lstB = lstA
lstA[1] = 10

print(lstB) # modified by the change in lstA

lstB[2] = 30

print(lstA) # modified by the change in lstB
````

If you are coming from a language like Matlab, you might not have expected editing the list represented by each variable lstA and lstB to affect the other. However, when Python evaluates the expression lstB = lstA, it does not create a copy of the original list and assign it to the new variable, instead *both variables are linked to the same list*. This means that manipulation of either variable affects the other (this is one reason why immutable objects like tuples can be useful for minimising the opportunites for errors to creep in unexpectedly).

When we do need to create a separate instance of a mutable object like a list, we instead use the copy() function:

````python
lstA = [1,2,3,4]
lstB = lstA.copy() # by copying the object, we do not modigy the original anymore
lstA[1] = 10

print(lstB) # now lstB is not affected by a change in lstA

lstB[2] = 30

print(lstA) # now lstA is not affected by a change in lstB
````

In NumPy, the syntax is slightly different, but the basic idea is the same:

````python
arA = np.array([1,2,3,4])
arB = arA
arC = np.copy(arA)

arB[2] = 20

print(arA)
print(arB)
print(arC)
````
