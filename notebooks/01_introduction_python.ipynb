# Objects in Python

Python is based on object-oriented programmming (OOP).Eeverything in Python is an object instead of just functions. Each object has data (attributes) such as identity, a type, and a value and behaviour (methods/fucntions) bundled together. There are different types of bjects in Python.

## 1. Numeric  
These represent numbers and are generally immutable
1.  int: Unlimited precision integers (e.g., 10, -5)
2. float: Double-precision floating-point numbers (e.g., 3.14)
3. complex: Numbers with real and imaginary parts (e.g., 2+3j)

## 2. Sequences
These store collections of objects in a specific order.
1. str: Immutable sequences of Unicode characters.
2. list: Mutable sequences used for collections of items.
3. tuple: Immutable sequences, often used for fixed records.
4. range: Immutable sequences of numbers typically used in loops.

## 3. Mapping and Sets
These handle unordered collections and unique data. 
1. dict: Mappings of unique keys to values (mutable).
2. set: Unordered collections of unique, hashable objects (mutable).
3. frozenset: An immutable version of a set. 



4. Truth and Null Types: 
bool: Represents True and False; it is a subclass of integers.
NoneType: Represents the absence of a value, with the single object None

5. Callable and Internal Objects:
These objects "do" things or manage execution. 
Functions: Both user-defined (def) and built-in (e.g., len()).
Classes & Instances: Blueprints and the individual objects created from them.
Methods: Functions bound to a class instance.
Modules: Objects that represent a file of Python code.
Code Objects: Compiled bytecode used by the interpreter.

6. Binary Sequence Types:
Used for manipulating raw bytes of data. 
bytes: Immutable arrays of single bytes.
bytearray: Mutable version of the bytes object.
memoryview: Allows access to an object's internal data without copying.

Among their main differences are whether an object are mutable (i.e., you can change them or not) or not: 
Inmutable: int, float, complex, str, tuple, frozenset, bytes	
Mutable: list, dict, set, bytearray

# Create your own class

In Python, a class is a blueprint or template for creating objects. It allows you to group data (attributes) and behaviours (methods) into a single logical unit. For example, a "Car" class might specificy that every car has a color and can drive(). The constructor __init__ is an special method that runs automatically when you create a new object. Its main job is to set the initial state (attrributes) of the object. 
The self parameter is mandatory first argument for methods within a class. It refers to the specific instance of the object being worked on, allowing Python to distinguish between differenrt objects of the same class.  

Use Class for complex workflows

```python
# Define class Bacteria with attributes (name, shape, gram_stain) and method (describe itself))
class Bacteria:
    microorganism_type = "Prokaryote"
    def __init__(self, name, shape, gram_stain):
        self.name = name
        self.shape = shape
        self.gram_stain = gram_stain
    def describe(self):
        return f"{self.name} is a {self.shape}-shaped, {self.gram_stain} bacterium."
```

```python
my_bacteria = Bacteria("Escherichia coli", "rod", "Gram-negative")
print(my_bacteria.describe())
```
