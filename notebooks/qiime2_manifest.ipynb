# From raw microbiome data to analysis-ready tables (ASVs) using QIIME2

## Goal
To transform IBDMDB data into a clean, structured, biologically meaningful data matrix, we will ise QIIME2 (https://qiime2.org/), which is an AI-ready microbiome multi-omics data science platform. QIMME2 will run in BASH (.sh) but we need a manifest.tsv file that contains the names of the samples and the path to the sequence file. We generate this manifest below using Python. 

```python
import numpy as np
import pandas as pd 
from pathlib import Path
import os
```

```python
# Paths setup

# Base path = where the project lives (works in local + Colab if consistent)
base_path = Path.cwd().resolve().parent

print("Current working directory:", base_path)

# Sequences directory
sequences_dir = base_path / "dataset" / "ibdmdb_hmp2" / "raw_sequences"
manifest_dir = base_path / "pipeline" / "qiime2" / "manifest"

# Change working directory
os.chdir(sequences_dir)

print("Now working in:", os.getcwd())

# Collect files
fastq_files = list(sequences_dir.glob("*.fastq.gz"))

# Build manifest
manifest = pd.DataFrame({
    "sample-id": [f.stem.replace(".fastq", "") for f in fastq_files],
    "absolute-filepath": [str(f.resolve()) for f in fastq_files]
})

# Optional: sort for reproducibility
manifest = manifest.sort_values("sample-id")

#Make output directory for manifest file
os.makedirs(manifest_dir, exist_ok=True) # Create manifest directory for downstream analyses

os.chdir(manifest_dir) # Change to manifest directory for downstream analyses

# Save to manifest file
manifest.to_csv("manifest.tsv", sep="\t", index=False)

manifest.head()
```
